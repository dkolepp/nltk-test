# Overview

Import basic Natural Language Toolkit libraries.
Going through the NLTK book: https://www.nltk.org/book/ch01.html


```python
import nltk
from nltk.book import FreqDist
from nltk.text import Text
import os
%matplotlib inline
```

Set variables for the notebook:

```python
RHT_XML_CORPUS_PATH='/mnt/VMs/virtualenv/nltkenv/share/nltk_data/corpora/redhat-training'
```

```python
file_ids = [x for x in os.listdir(RHT_XML_CORPUS_PATH) if x.endswith('.xml')]
```

```python
rht_corpus = nltk.corpus.reader.xmldocs.XMLCorpusReader(RHT_XML_CORPUS_PATH,
                                                        file_ids)
```

The `words()` method generates a list of words from the corpus.

```python
rht_corpus.words()
```

The `Text` class provides lots of handy methods for examining a text.

```python
do180 = Text(rht_corpus.words())
```

The `concordance` method displays instances of the word with context.

```python
word_token = 'OpenShift'
do180.concordance(word_token)
```

```python
do180.similar(word_token)
```

Dispersion plot of words

```python
do180.dispersion_plot(['Container', 'OpenShift', 'Podman'])
```

## Frequency Distributions

```python
fdist1 = FreqDist(do180)
print(fdist1)
```

Most Common:

```python
fdist1.most_common(50)
fdist1.plot(50, cumulative=True)
```
